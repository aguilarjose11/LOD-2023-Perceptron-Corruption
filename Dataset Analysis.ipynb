```python
import pandas as pd
import numpy as np
from scipy.io import arff

import os
import pickle
```

# SPECT Heart Data Set

```python
SPECT_folder = './SPECT/'
SPECT_train_file  = os.path.join(SPECT_folder, 'SPECTF.train')
SPECT_test_file   = os.path.join(SPECT_folder, 'SPECTF.test')
%ls SPECT
```

```python
SPECT_train = pd.read_csv(SPECT_train_file, header=None)
SPECT_test  = pd.read_csv(SPECT_test_file, header=None)
SPECT       = pd.concat((SPECT_train, SPECT_test), axis=0).reset_index(drop=True)
```

```python
SPECT_train
```

```python
SPECT[0].value_counts()
```

```python
SPECT.describe()
```

## Storing Data as Pickle File

```python
X = SPECT.drop(0, axis=1)
y = SPECT[0]

cucumber = {
    'X': X,
    'y': y,
}

with open('SPECT.pkl', 'wb') as jar:
    pickle.dump(cucumber, jar)
```

# Polish Bankrupcy Dataset

```python
PBankrupcy_folder = './PolishBankrupcy/'
#PBankrupcy_file = os.path.join(PBankrupcy_folder, 'data.csv')
%ls PolishBankrupcy
```

```python
PBankrupcy_1 = pd.DataFrame(arff.loadarff(os.path.join(PBankrupcy_folder, '1year.arff'))[0])
PBankrupcy_1['class'] = PBankrupcy_1['class'].astype('int32')
PBankrupcy_2 = pd.DataFrame(arff.loadarff(os.path.join(PBankrupcy_folder, '2year.arff'))[0])
PBankrupcy_2['class'] = PBankrupcy_2['class'].astype('int32')
PBankrupcy_3 = pd.DataFrame(arff.loadarff(os.path.join(PBankrupcy_folder, '3year.arff'))[0])
PBankrupcy_3['class'] = PBankrupcy_3['class'].astype('int32')
PBankrupcy_4 = pd.DataFrame(arff.loadarff(os.path.join(PBankrupcy_folder, '4year.arff'))[0])
PBankrupcy_4['class'] = PBankrupcy_4['class'].astype('int32')
PBankrupcy_5 = pd.DataFrame(arff.loadarff(os.path.join(PBankrupcy_folder, '5year.arff'))[0])
PBankrupcy_5['class'] = PBankrupcy_5['class'].astype('int32')
PBankrupcy = pd.concat((PBankrupcy_1, PBankrupcy_2, PBankrupcy_3, PBankrupcy_4, PBankrupcy_5), axis=0)
```

```python
PBankrupcy
```

```python
PBankrupcy.describe()
```
